GreyScaling

```python
x=1
c=1.2
for x in range(1,54):
    
    if(x<10):
        path="Project input/IDRiD_0"+str(x)+".jpg"
    else:
        path="Project input/IDRiD_"+str(x)+".jpg"
    img=cv2.imread(path,cv2.IMREAD_GRAYSCALE)
    img2=cv2.imread(path,cv2.IMREAD_GRAYSCALE)
    print(img.shape)
    print(x)
    h=img.shape[0]
    w=img.shape[1]
    imax=255
    for i in np.arange(h):
        for j in np.arange(w):
            a=img.item(i,j)
            e=math.ceil(a*c)
            if(e>255):
                e=255
            img2.itemset((i,j),e)
            b=imax-e
            img.itemset((i,j),b)

    img_r=cv2.resize(img,(250,250))  
    path2="GreyScaled Project Output/IDRiD_000"+str(x)+".jpg"   #inverted
    cv2.imwrite(path2,img_r)
    img_r1=cv2.resize(img2,(250,250))
    path3="GreyScaled Project Output1/IDRiD_000"+str(x)+".jpg"   #Contrast
    cv2.imwrite(path3,img_r1)
```
